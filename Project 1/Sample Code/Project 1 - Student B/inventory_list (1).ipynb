```python
def load_inventory(player):
    filename1 = 'weapons.json'
    filename2 = 'armors.json'
    with open(filename1) as w:
        weapon_list = json.load(w)
    with open(filename2) as w:
        armor_list = json.load(w)
    for items in weapon_list:
            weapon = Weapon(items["name"], items["description"], items["slot"], items["damage"],
                        items["bonus_attack"], items["bonus_defense"],items["bonus_magic"],
                        items["max_hp"])
            player.add_to_inventory(weapon)
    for items in armor_list:
            armor = Armor(items["name"], items["description"], items["slot"], items["damage"],
                        items["bonus_attack"], items["bonus_defense"],items["bonus_magic"],
                        items["max_hp"])
            player.add_to_inventory(armor)
    
```
